```python
#import required libraries
import anonypy
import pandas as pd
```

```python
data = pd.read_csv("adult.csv")
data.columns = data.columns.str.strip()

cols = set(("workclass", "education", "marital.status", "occupation", "sex", "native.country"))
for col in cols:
    data[col] = data[col].astype("category")
```

```python
quasi_identifiers = ["age", "education", "marital.status", "occupation", "native.country"]
sensitive_attributes = "income"
```

```python
p = anonypy.Preserver(data, quasi_identifiers, sensitive_attributes)
rows = p.anonymize_k_anonymity(k=10)

dfn = pd.DataFrame(rows)
dfn.head()
```
